How are loan status and employment status related?

```python
#exploring employment status vs loan status
status_emp = target.groupby(['LoanStatus', 'EmploymentStatus']).size().reset_index(name='count')
sns.catplot(x='LoanStatus', y='count', hue='EmploymentStatus', data=status_emp, kind='bar', height=5, aspect=2)
plt.title('Employment Status vs. Loan Status')
plt.xlabel('Loan Status', rotation = 90)
plt.ylabel('Count of Loans')
plt.show()
```

```python
#let us make that clearer using seaborn
status_emp = target.groupby(['LoanStatus', 'EmploymentStatus']).size().reset_index(name='count')
ax = sns.catplot(x='LoanStatus', y='count', hue='EmploymentStatus', data=status_emp, kind='bar', height=5, aspect=2)
ax.set_xticklabels(rotation=90)
plt.title('Employment Status vs. Loan Status')
plt.xlabel('Loan Status')
plt.ylabel('Count of Loans')
plt.show()
```

Most current loans are representing employed loanees, and most who have completed are employed too

How does the borrower APR affect the loan amount?

```python
#loan amount vs borrower APR

sns.scatterplot(x='LoanOriginalAmount', y='BorrowerAPR', data=target, alpha = 0.3, s =4)
plt.title('Loan Amount vs. Borrower APR')
plt.xlabel('Loan Amount')
plt.ylabel('Borrower APR')
plt.show()

```

From the scatter plot, we can see that there is a weak positive relationship between loan amount and borrower's APR. As loan amount increases, borrower's APR tends to increase as well, but the relationship is not very strong. However, we can also see that there are some outliers where larger loans have lower APRs than expected.

How do loan status and term affect Credit Grade?

```python

#heat map to see the relationship of  Loan Status, Loan Term, and Credit Grade
status_term_grade = df.groupby(['LoanStatus', 'Term', 'CreditGrade']).size().reset_index(name='count')
pivot = status_term_grade.pivot_table(index=['Term', 'CreditGrade'], columns='LoanStatus', values='count', fill_value=0)
plt.figure(figsize=(10, 5))
sns.heatmap(pivot, cmap='Blues', annot=True, fmt='d')
plt.title('Loan Status, Loan Term, and Credit Grade')
plt.xlabel('Loan Status')
plt.ylabel('Loan Term, Credit Grade')
plt.show()
```

From the heatmap, we can see that most loans are current, followed by completed loans. We can also see that most loans have a term of 36 months, with fewer loans having a term of 60 months. In terms of credit grade, most loans have a credit grade of C or D, with fewer loans having a credit grade of A or AA. The heatmap also shows that loans with a term of 60 months and a credit grade of A or AA are relatively rare.
